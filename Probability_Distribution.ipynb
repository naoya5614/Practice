# 確率分布

```python
#各種モジュールの読み込み
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import numpy as np

#seabornのスタイルを設定
sns.set()
```

## 二項分布

```python
# 二項分布 試行回数：100, 確率：0.3, サンプル数：10000）
r_binom = np.random.binomial(n=100, p=0.3, size=10000)
x, count = np.unique(r_binom, return_counts = True) 
plt.bar(x, count)
print('平均値', np.mean(r_binom))
print('分散', np.var(r_binom, ddof=0))
```

## ポアソン分布

```python
#ポアソン分布 random.poisson(lam=1.0, size=None)
r_pois = np.random.poisson(size= 1000, lam=5)
x, count = np.unique(r_pois, return_counts = True) 
plt.bar(x, count)
print('平均値', np.mean(r_pois))
print('分散', np.var(r_pois, ddof=0))
```

## 正規分布

```python
# 正規分布
# np.random.normal(平均：50, 標準偏差：20, サンプル数：1000)
r_norm = np.random.normal(loc=50, scale=20, size=1000)
sns.displot(r_norm)
print('平均値', np.mean(r_norm))
print('分散', np.var(r_norm, ddof=0))

# np.random.randn 平均：50, 標準偏差：20, サンプル数：1000
r_norm2 = 20 * np.random.randn(1000) + 50
sns.displot(r_norm2)
print('平均値', np.mean(r_norm2))
print('分散', np.var(r_norm2, ddof=0))
```

```python
# scipyで正規分布を描写
from scipy import stats

# 1〜100までの数字を0.1刻み
nx = np.arange(start=1, stop=100.1, step=0.1)

#確率密度の計算
norm_d = stats.norm.pdf(x = nx, loc=50, scale=10) #loc平均 scale標準偏差

#分布の描画
plt.plot(nx, norm_d, color='green')
```

```python
norm_d = stats.norm.pdf(x = nx, loc=50, scale=20) #loc平均 scale標準偏差

#分布の描画
plt.plot(nx, norm_d, color='green')
```

## カイ二乗分布

```python
#カイ二乗分布 np.random.chisquare(df=自由度, size=サンプル数)

#自由度1
r_chi = np.random.chisquare(df=1, size=1000)
sns.displot(r_chi)

#自由度5
r_chi2 = np.random.chisquare(df=5, size=1000)
sns.displot(r_chi2)

#自由度10
r_chi3 = np.random.chisquare(df=10, size=1000)
sns.displot(r_chi3)
```

```python
# 　一様分布 0.1以上、1.0未満

#サイズが1
r_ran = np.random.rand(1)
sns.displot(r_ran)

#サイズが10
r_ran2 = np.random.rand(10)
sns.displot(r_ran2)

#サイズが5×5
r_ran3 = np.random.rand(5, 5)
sns.displot(r_ran3)

#  numpy.random.random_sample(): 0.0以上、1.0未満
#  numpy.random.randint(): 任意の範囲の整数
```

## 指数分布

```python
# 指数分布
#標準偏差1 サンプル数1000
r_expo = np.random.exponential(1, 1000)
sns.displot(r_expo)
```

## ベータ分布

```python
# ベータ分布

#α=2, β=5, サンプル数1000
r_beta = np.random.beta(2, 5, 1000)
sns.displot(r_beta)
```

## ガンマ分布

```python
# ガンマ分布

#形状母数 k=3、尺度母数 θ=1, サンプル数1000
r_gamma = np.random.gamma(3, 1, 1000)
sns.displot(r_gamma)
```
